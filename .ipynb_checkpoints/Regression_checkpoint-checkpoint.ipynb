```python
import matplotlib.pyplot as plt
import seaborn as sns
#which features that are important or not important to the price
#link betwwen sqft_living and sqft_living15, same for: sqft_lot and sqft_lot15
#show outliers for the iportant features

def plot_correlation_map( df, m ):
    corr = df.corr()
    s , ax = plt.subplots( figsize =( m , m ) )
    cmap = sns.diverging_palette( 256, 0 , as_cmap = True )
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 10 }
        )
```

```python
# first: which features that are important or not important to the price
#Try running the function, analyse what does it do exactly and what are the utilities of it,
plot_correlation_map( df, 12 )

# from the following graph the id is clearly unsignificant feature
```

```python
df_new= df.drop('id', axis='columns', inplace=False)
df_new= df_new.drop('zipcode', axis='columns', inplace=False)
plot_correlation_map( df_new, 10 )

#sqft_living and sqft_lot have a high correlations with sqft_living15 and sqft_lot15 
#but sqft_living and sqft_lot have the highest impact on the price
```

```python
df_new= df_new.drop('sqft_living15', axis='columns', inplace=False)
df_new= df_new.drop('sqft_lot15', axis='columns', inplace=False)
plot_correlation_map( df_new, 10 )

#from the correlation graph, it is preffered to do a regression on (all the rest features) 
#and an other regression on (bedrooms, bathrooms, sqrt_living, grade, sqrt_above, view)
```

```python
#linear regression on df_all
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics

#same for df_all and df_small

#extract x and y from our data
x=df_all["sqft_living"].values[:,np.newaxis]
y=df_all["price"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%

model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance

#error measurement
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))

plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
#polynomial regression with outliers on these features: 
#'price','bedrooms','bathrooms', 'sqft_living', 'grade', 'sqft_above', 'sqft_basement','view'

from sklearn.preprocessing import PolynomialFeatures 

from sklearn.metrics import mean_squared_error
from sklearn import metrics

x= df_small[['sqft_living', 'grade']]
y= df_small["price"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  #splitting data
lg=LinearRegression()
poly=PolynomialFeatures(degree=3)

x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)

print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```
