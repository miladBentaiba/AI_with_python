```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
dataset
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_).astype(int)  #Creating a new DataFrame from our Numpy array
df.head()
```

```python
df['sum']=df.sum(axis=1)
df.head()
```

```python
#Let’s select itemsets with a minimum of 60% Support
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
#Support with column names
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
#confidence code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
#Lift code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules

dataset = pd.read_csv("Market_Basket_Optimisation.csv",header=None)
dataset.head()
```

```python
products = pd.DataFrame(dataset[0].value_counts())
best_20_product = pd.DataFrame(dataset[0].value_counts()).head(20)

plt.figure(figsize=(12,10))

sns.barplot(best_20_product.index,best_20_product[0])

plt.title('20 most popular products vs. amount', fontsize = 20)
plt.xlabel('Most popular products', fontsize = 15)
plt.ylabel('Amount', fontsize = 15)

plt.xticks(rotation = '90')
plt.show()
```

```python
#clean content
dataset=dataset.apply(lambda x: x.str.strip())
dataset.fillna('',axis=1,inplace=True)

#transform dataset
te=TransactionEncoder()
df_out = dataset.to_numpy().tolist()
te_ary=te.fit(df_out).transform(df_out)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_).astype(int)  #Creating a new DataFrame from our Numpy array
df.columns
```

```python
#drop nan columns
df.drop('', axis='columns', inplace=True)
df.columns
```

```python
#Let’s select itemsets with a minimum of 10% Support
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.08)
```

```python
#Support with column names
frequent_itemsets=apriori(df, min_support=0.02, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
#confidence code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence", min_threshold=0.3) # associate itemsets with confidence over 70%.
```

```python
#Lift code
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.5)
```
